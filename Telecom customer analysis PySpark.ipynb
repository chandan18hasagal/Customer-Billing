```python
## 21 Define the SparkMl Pipelines for the preprocessing stages
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
import numpy as np
import StringIO
import pandas as pd
import warnings
```

```python
## Combine all the numerical columns as vector 
## i) Using VectorAssembler to combine
num_Var_Names = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
vector_assembler_NumVars = VectorAssembler(inputCols=num_Var_Names, 
                                           outputCol="num_features_all")
```

```python
## ii) Scaling all the numeric attributes using MinMaxScaler
```

```python
from pyspark.ml.feature import MinMaxScaler
```

```python
min_Max_Scalar_NumVars = MinMaxScaler(inputCol="num_features_all", 
                                      outputCol = "scaled_num_features")
```

```python
cat_Var_Names = ['Contract', 'Dependents', 'DeviceProtection', 'InternetService', 'MultipleLines', 
                 'OnlineBackup', 'OnlineSecurity', 'PaperlessBilling', 'Partner','PhoneService','SeniorCitizen','StreamingMovies','StreamingTV','TechSupport','gender','PaymentMethod']
```

```python
## iii)  StringIndexer and OneHotEncoderEstimator for the categorical columns as a single vector
from pyspark.ml.feature import OneHotEncoder, StringIndexer
```

```python
indexers_Cat  = [StringIndexer(inputCol=cat_Var_Name, 
                               outputCol="{0}_index".format(cat_Var_Name))
                 for cat_Var_Name in cat_Var_Names ]
```

```python
encoders_Cat  = [OneHotEncoder(inputCol=indexer.getOutputCol(), 
                               outputCol="{0}_vec".format
                               (indexer.getInputCol())) 
                 for indexer in indexers_Cat]
```

```python
## iv)   Combine all the vectors created by OneHotEncoderEstimator
assembler_Cat = VectorAssembler(inputCols=[encoder.getOutputCol() 
                                           for encoder in encoders_Cat], 
                                outputCol="cat_features")
```

```python
## v) Combine both the scaled numerical vector and combined categorical vectors 
assembler     = VectorAssembler(inputCols=["scaled_num_features","cat_features"], 
                                outputCol="features")
```

```python
## Complete Pipiline
```

```python
preprocessiong_Stages = [vector_assembler_NumVars] +\
                        [min_Max_Scalar_NumVars] +\
                        indexers_Cat + encoders_Cat +\
                        [assembler_Cat] +\
                        [assembler]
```
